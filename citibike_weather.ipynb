```python
url = 'http://api.openweathermap.org/data/2.5/air_pollution'
```

```python
df = pd.read_csv('Resources/citibike_stations.csv')
```

```python
df['aqi'] = ''
```

```python
df
```

```python
counter = 1
set_count = 1
for index, row in df.head(500).iterrows():
    
    print(index)
    
    # set params
    myparams = {
        'lat':row['lat'],
        'lon':row['lon'],
        'appid':apikey
    }
    
    # make req
    req = requests.get(url, params=myparams)
    resp = req.json()
    
    try:
        # get aqi
        aqi = resp['list'][0]['main']['aqi']

        # set value in df
        df.loc[index, 'aqi'] = aqi
    
    # skip if key not found
    except KeyError as e:
        pass
    
    counter += 1
    print('-'*10)
    time.sleep(1)
```
